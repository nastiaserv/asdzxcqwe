```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Fills and Alpha Example")
plt.show()
```

```python
import os
parent_folder_path = '/content/parent_folder'
if not os.path.exists(parent_folder_path):
    os.makedirs(parent_folder_path)
    print(f'Створено нову папку: {parent_folder_path}')
else:
    print(f'Папка {parent_folder_path} вже існує.')
```

```python
import os
parent_folder_path = '/content/parent_folder'
new_folder_name = 'child_folder'
new_folder_path = os.path.join(parent_folder_path, new_folder_name)
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)
    print(f'Створено нову папку: {new_folder_path}')
else:
    print(f'Папка {new_folder_path} вже існує.')
```
